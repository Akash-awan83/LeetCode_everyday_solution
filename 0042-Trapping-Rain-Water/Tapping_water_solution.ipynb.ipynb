# Trapping Rain Water
Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

# Example
Input: height = [4,2,0,3,2,5]
Output: 9
 

# Constraints:

n == height.length
1 <= n <= 2 * 104
0 <= height[i] <= 105

# BruteForce Approach

**In this Approch we have a timecomplexity of O(n)^2**

**And Space Complexity of O(1)**


```python
class Solution:
    def trap(self,arr):
        n=len(arr)
        total_water=0
        #we will skip first and last one because they can't store water
        for i in range(1,n-1):
            #now we will move from right to left and find left max 
            left_max=0
            for j in range(i,-1,-1):
                left_max=max(left_max,arr[j])
            #Now we will move from  left to right and find right max
            right_max=0
            for j in range(i,n):   
                right_max=max(right_max,arr[j])
            #here is our total water   
            total_water+=min(left_max,right_max)-arr[i]
        return total_water    
    
    
obj=Solution()    
arr=[4,2,0,5,2,6,2,3]
print(f"output is : {obj.trap(arr)}")
```

# Optimal Approach Two Pointer Approach
**In this Approach we have time complexity of O(N)**
**And space complexity of O(1)**

```python
class Solution():
    def trap(self,nums):
        left_max=0
        right_max=0
        n= len(nums)
        ans=0
        left=0
        right=n-1
        while left<=right:
            if left_max<=right_max:
                left_max=max(left_max,nums[left])
                ans+=left_max-nums[left]
                left+=1
            else:
                right_max=max(right_max,nums[right]) 
                ans+=right_max-nums[right] 
                right-=1    
        return ans    
            
            
L=Solution()       
nums=[5,2,0,5,2,4]
print(L.trap(nums))       
```
