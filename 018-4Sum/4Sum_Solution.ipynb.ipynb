# 4Sum

Given an array nums of n integers, return an array of all the unique quadruplets [nums[a], nums[b], nums[c], nums[d]] such that:

0 <= a, b, c, d < n
a, b, c, and d are distinct.
nums[a] + nums[b] + nums[c] + nums[d] == target
You may return the answer in any order.

Example 1:

Input: nums = [1,0,-1,0,-2,2], target = 0
Output: [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]
 

# BruteForse Approach

```python
class Solution:
    def FourSum(self,nums,target):
        temp=set()
        n=len(nums)
        for i in range(n):
            for j in range(i+1,n):
                for k in range(j+1,n):
                    for l in range(k+1,n):
                        sum=nums[i]+nums[j]+nums[k]+nums[l]
                        if sum==target:
                            tup=tuple(sorted([nums[i],nums[j],nums[k],nums[l]]))
                            temp.add(tup)
        return list(temp)                    
```

```python
tp=Solution()
nums = [1,0,-1,0,-2,2] 
target = 0
print(tp.FourSum(nums,target))
# print(l)

```

# Better then Brute Force

```python
from collections import defaultdict
class Solution:
    def fourSum(self,nums,target):
        n=len(nums)
        temp=set()
        hmap=defaultdict(int)
        for i in  nums:
            hmap[i]+=1
        for i in range(n-3):
            hmap[nums[i]]-=1
            for j in range(i+1,n-2):
                hmap[nums[j]]-=1
                for k in range(j+1,n-1):
                    hmap[nums[k]]-=1
                    rem=target-(nums[i]+nums[j]+nums[k])
                    if rem in hmap and hmap[rem]>0:
                        tup=tuple(sorted([nums[i],nums[j],nums[k],rem]))
                        temp.add(tup)
                    hmap[nums[k]]+=1
                hmap[nums[j]]+=1
            hmap[nums[i]]+=1
                              
        return list(temp)            
                   
            
```

```python
tp=Solution()
nums = [1,0,-1,0,-2,2] 
target = 0
print(tp.fourSum(nums,target))
# print(l)
```

# Optimal Approach

```python
class Solution:
    def fourSum(self,nums,target):
        res=[]
        nums.sort()
        n=len(nums)
        for i in range(n-3):
            #avoiding duplicates for i
            if i > 0 and nums[i]==nums[i-1]:
                continue
            for j in range(i+1,n-2):
                #avoiding duplicates for j
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue
                p=j+1
                q=n-1
                while p<q:
                    sum=nums[i]+nums[j]+nums[p]+nums[q]
                    if sum==target:
                        res+=[[nums[i],nums[j],nums[p],nums[q]]]
                        # skip duplicates
                        while p < q and nums[p] == nums[p + 1]:
                            p+= 1
                        p+= 1
                        while p < q and nums[q] == nums[q - 1]:
                            q-= 1

                    elif sum>target:
                        q-=1
                    else:
                        p+=1        
        return res   
 
# print(l)            
```

```python
tp=Solution()
nums = [1,0,-1,0,-2,2] 
target = 0
print(tp.fourSum(nums,target))
```
